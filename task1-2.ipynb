## 1) Byte Pair Encoding (BPE) Algorithm

BPE was first developed for **text compression tasks**, but later adapted for **tokenization in transformer models**.  

The core idea of BPE is to **iteratively merge the most frequent pairs of symbols (characters or subwords) into new tokens**.

---

### Steps of the Algorithm
1. **Normalization & Pre-tokenization**  
   Clean the text (lowercasing, Unicode normalization, etc.) and split it into initial tokens (usually characters).

2. **Initialize Vocabulary**  
   Compute all unique characters in the corpus and create the initial vocabulary.

3. **Iterative Merging**  
   - Count the frequency of all adjacent token pairs.  
   - Select the most frequent pair.  
   - Merge it into a new token and add it to the vocabulary.  
   - Repeat until the vocabulary reaches the desired size.

---

### Example
Suppose our corpus only contains these words: 

look,book,took

**Step 1: Initial vocabulary (unique characters)**:

t, o, l, b, k

**Step 2: First merge (most frequent pair is "oo")**:

t, o, l, b, k, oo, ook

**Step 3: Next merge (most frequent is "ook")**:

t, o, l, b, k, oo, ook

This process continues until the vocabulary reaches the desired length.

---

✅ In summary, BPE starts with characters, then keeps merging the most frequent symbol pairs into larger subwords, building a vocabulary that balances efficiency and coverage.

### *Which LLMs use BPE?*

   LLMs like GPT,GPT-2,RoBERTa,BART,DeBERTa use BPE tokenizer


```python
from transformers import AutoTokenizer
import pprint
tokenizer = AutoTokenizer.from_pretrained("gpt2")
print(f"GPT-2's vocab size is : {tokenizer.vocab_size}")
text = "hey this is project-404 groups's nlp task"
text_per = "سلام این تسک گروه پروجکت 404 هست"
ids_per = tokenizer.encode(text_per)
ids = tokenizer.encode(text)
print(ids)
print(tokenizer.decode(ids))
print(tokenizer.decode(ids_per))
tokens_per = ids = tokenizer.tokenize(text_per)
tokens = ids = tokenizer.tokenize(text)
print(tokens_per)
print(tokens)
pprint.pprint(tokenizer.vocab)
```

```python
from transformers import RobertaTokenizer
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")
print(f"RoBERTa's vocab size is : {tokenizer.vocab_size}")
text = "hey this is project-404 groups's nlp task"
text_per = "سلام این تسک گروه پروجکت 404 هست"
ids = tokenizer.encode(text)
ids_per = tokenizer.encode(text_per)
print(f"IDS of tokens that were used in English text in RoBERTa's vocab:{ids}")
print(f"IDS of tokens that were used in Persian text in RoBERTa's vocab:{ids_per}")
print(tokenizer.decode(ids))
print(tokenizer.decode(ids_per))
tokens_per = tokenizer.tokenize(text_per)
tokens = tokenizer.tokenize(text)
print(tokens_per)
print(tokens)
pprint.pprint(tokenizer.get_vocab())
```
